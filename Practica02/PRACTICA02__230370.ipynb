12.3 funciones que mandan llamar optras funiones y que regresan el resultado en un procedimiento en estecaso utilizaremos una funcion predefinida en denominada replacve que nos permiten sustituir letras en unacadena de texto

```python
def sustituir_vocales(texto):
    texto = texto.replace("A", "4")
    texto = texto.replace("E", "3")
    texto = texto.replace("I", "1")
    texto = texto.replace("O", "0")
    texto = texto.replace("U", "∪")

    texto = texto.replace("a", "4")
    texto = texto.replace("e", "3")
    texto = texto.replace("i", "1")
    texto = texto.replace("o", "0")
    texto = texto.replace("u", "∪")
    return texto
```

18.- **Eliminar los datos de cuartiles menores**

Con la finalidad de hacer un analisis centrado en los datos tendencia es decir con mayor frecuencia eliminamos los registros que entran en el cuartil menor a 0.05 y mayor a 0.95
calculamos un límite superior,
calculamos un límite inferior,
y filtramos únicamente los valores que se encuentran dentro de un rango considerado normal.
El método quantile() permite calcular percentiles de una columna numérica.


```python
a=df['MILES'].quantile(0.95)
b=df['MILES'].quantile(0.05)
df3=df[df['MILES']>b]
df3=df[df['MILES']<a]
df3
```

```python
df3['MILES'].describe()
```

19.- **calculo de desviacion estandar**

para entender que tan variable o dispersos son las distancias de los viajes

```python
df3['STD']=df.MILES.std()
df3
```

20.- **Estimar el valor de Z-Score p puntuacion z**

con la finalidad de estimar la dispersion de los datos, es decire que tan variado o diferente son, utilizamos eesta formula estadistica para poder continuar con el analisis EDA (Exploratory Data Analisys)


```python
df3['Z']=(df.MILES-df.MILES.mean())/df.MILES.std()
df3
```

21.- Filtramos aquello que no este en el rango -1 y 1

```python
df3=df3[(df3.Z>-1) & (df3.Z<1)]
df3
```

25.**Grafico de lugar de abordaje (START)**

```python
fig, ax = mat.subplots(figsize=(15,50))

orden = sorted(df['START'].dropna().unique())

grafica = sea.countplot(
    data=df,
    y='START',
    order=orden,
    hue='START',
    dodge=False,
    palette='pastel',
    legend=False,
    ax=ax

)

ax.set_title('Lugarde abordaje')

for container in ax.containers:
    ax.bar_label(container)

mat.show()
```
